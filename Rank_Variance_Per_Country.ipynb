```python
# Import your libraries
import pandas as pd

# Start writing code
#fb_comments_count
#fb_active_users

#fb_comments_count.shape[0]

fb_comments_count_filtered_in_2019 = fb_comments_count[(fb_comments_count['created_at'] >= '2019-12-01') &
(fb_comments_count['created_at']<='2020-01-01')]
#fb_comments_count_filtered_in_2019.shape[0]


result_2019 = fb_comments_count_filtered_in_2019.groupby('user_id')['number_of_comments'].sum().reset_index()
result_2019.columns = ['user_id','total_comments_in_2019']
#result

final_df_2019 = fb_active_users.merge(result_2019, on = 'user_id')
#final_df

total_country_count_in_2019 = final_df_2019.groupby('country')['total_comments_in_2019'].sum().reset_index()
#total_country_count_in_2019.columns = ['country','total_count_of_comments']


total_country_count_in_2019=total_country_count_in_2019.sort_values('total_comments_in_2019',ascending=False)
total_country_count_in_2019['rank_in_2019'] = total_country_count_in_2019['total_comments_in_2019'].rank(ascending=False,method='dense')
total_country_count_in_2019

fb_comments_count_filtered_in_2020 = fb_comments_count[(fb_comments_count['created_at'] >= '2019-12-01') &
(fb_comments_count['created_at']<='2020-01-31')]

result_2020 = fb_comments_count_filtered_in_2020.groupby('user_id')['number_of_comments'].sum().reset_index()
result_2020.columns = ['user_id','total_comments_in_2020']
#result

final_df_2020 = fb_active_users.merge(result_2020, on = 'user_id')
#final_df

total_country_count_in_2020 = final_df_2020.groupby('country')['total_comments_in_2020'].sum().reset_index()
#total_country_count_in_2019.columns = ['country','total_count_of_comments']

total_country_count_in_2020=total_country_count_in_2020.sort_values('total_comments_in_2020',ascending=False)
total_country_count_in_2020['rank_in_2020'] = total_country_count_in_2020['total_comments_in_2020'].rank(ascending=False,method='dense')
total_country_count_in_2019
#total_country_count_in_2020

total_comparing_countries = total_country_count_in_2019.merge(total_country_count_in_2020, on='country')
total_comparing_countries['difference'] = total_comparing_countries['rank_in_2020']-total_comparing_countries['rank_in_2019']
total_comparing_countries=total_comparing_countries.sort_values('difference',ascending=True)

#total_comparing_countries.drop(['total_comments_in_2019','rank_in_2019','total_comments_in_2020','rank_in_2020'],axis=1,inplace=True)
output_df = total_comparing_countries.head(1)
output_df.drop(['total_comments_in_2019','rank_in_2019','total_comments_in_2020','rank_in_2020','difference'],axis=1,inplace=True)
output_df
```
