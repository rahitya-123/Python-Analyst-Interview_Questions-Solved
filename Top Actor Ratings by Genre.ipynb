```python
# Import your libraries
import pandas as pd

# Start writing code
#top_actors_rating.head()

# More explicit approach showing counts
def get_top_genre_with_rating_tiebreaker(group):
    genre_counts = group.groupby(['genre']).size()
    max_count = genre_counts.max()
    
    tied_genres = genre_counts[genre_counts==max_count].index
    
    if len(tied_genres)==1:
        return tied_genres[0]
        
    else:
        tied_data = group[group['genre'].isin(tied_genres)]
        avg_ratings =tied_data.groupby('genre')['movie_rating'].mean()
        return avg_ratings.idxmax()
        

top_genre_per_actor = top_actors_rating.groupby('actor_name').apply(get_top_genre_with_rating_tiebreaker)
#print(top_genre_per_actor)

# Create DataFrame directly
top_genre_per_actor_df = pd.DataFrame({
    'actor_name': top_genre_per_actor.index,
    'top_genre': top_genre_per_actor.values
})
#print(top_genre_per_actor_df)

top_genre_per_actor_data = top_genre_per_actor_df.merge(top_actors_rating, on='actor_name')
top_genre_per_actor_data = top_genre_per_actor_data[top_genre_per_actor_data['genre'] == top_genre_per_actor_data['top_genre']]

#print(top_genre_per_actor_data)
#top_genre_per_actor_data.drop('top_genre',axis=1,inplace = True)
#print(top_genre_per_actor_data.columns.tolist())

actor_avg_ratings = top_genre_per_actor_data.groupby(['actor_name','genre'])['movie_rating'].mean().reset_index()
actor_avg_ratings.columns = ['actor_name', 'genre', 'average_rating']

actor_avg_ratings = actor_avg_ratings.sort_values('average_rating',ascending=False)
#print(actor_avg_ratings)

actor_avg_ratings['rank'] = actor_avg_ratings['average_rating'].rank(method='dense', ascending=False)
#print(actor_avg_ratings)

top_3_actors = actor_avg_ratings[actor_avg_ratings['rank']<=3]
top_3_actors.drop('rank',axis=1,inplace=True)
top_3_actors.reset_index(drop=True,inplace=True)
top_3_actors
```
